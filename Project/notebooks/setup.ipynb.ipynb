```python
# Databricks notebook source
# Setup script for Medallion Architecture Pipeline (Bronze → Silver → Gold)

# Configuration - Update these values as needed
CATALOG_NAME = "main"  # Change to your catalog name
BRONZE_SCHEMA = "bronze"
SILVER_SCHEMA = "silver"
GOLD_SCHEMA = "gold"

print(f"Starting environment setup for catalog: {CATALOG_NAME}")
print("=" * 60)

# Step 1: Create catalog if not exists
print(f"\n1. Setting up catalog: {CATALOG_NAME}")
try:
    spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG_NAME}")
    print(f"   ✓ Catalog '{CATALOG_NAME}' is ready")
except Exception as e:
    print(f"   ⚠ Note: {e}")

# Step 2: Use the catalog
spark.sql(f"USE CATALOG {CATALOG_NAME}")
print(f"   ✓ Using catalog: {CATALOG_NAME}")

# Step 3: Create schemas for each layer
schemas = {
    BRONZE_SCHEMA: "Raw data layer - ingested CSV files",
    SILVER_SCHEMA: "Cleaned and validated data layer",
    GOLD_SCHEMA: "Business-level aggregates and analytics"
}

print(f"\n2. Setting up schemas:")
for schema_name, description in schemas.items():
    try:
        spark.sql(f"""
            CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{schema_name}
            COMMENT '{description}'
        """)
        print(f"   ✓ Schema '{schema_name}' is ready")
    except Exception as e:
        print(f"   ⚠ Error creating schema '{schema_name}': {e}")

# Step 4: Verify setup
print(f"\n3. Verification:")
print(f"   Current catalog: {spark.sql('SELECT current_catalog()').collect()[0][0]}")
```
